##  Un peu de pre-processing avec la team Star Wars  

Importer les 10 premiers personages de star wars depuis l'api du site [swapi](https://swapi.co/) et les ranger dans un dataframe de la√ßon suivante : 
<img src="./StarWarsResult.png">

```python
import requests
import pandas as pd
```

```python
r = requests.get('https://swapi.co/api/people/1')
```

```python
data = r.json()
```

```python
df = pd.DataFrame(columns=data.keys())
columns = list(df)
```

```python
df
```

```python
r = requests.get('https://swapi.co/api/people/1')
data = r.json()
data
```

```python
type(data["films"])
```

```python
for x in range(1, 11):
    r = requests.get('https://swapi.co/api/people/' + str(x))
    data = r.json()
    for key in columns:
        df.loc[x, key] = str(data[key]).strip('[]')
df
```
