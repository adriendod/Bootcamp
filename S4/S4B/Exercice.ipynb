```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
expo = np.random.exponential(size=10000000)
```

```python
sns.distplot(expo)
```

```python
list = np.split(expo, 1000)
list = np.asarray(list)
```

```python
list.shape
```

```python
mean = np.mean(list, axis = 1)
```

```python
sns.distplot(mean)
```

```python
np.mean(mean)
```

```python
np.std(mean)
```

### Loi normale

```python
normal = np.random.normal(np.mean(mean), np.std(mean), 1000)
```

```python
sns.distplot(normal, hist = False)
sns.distplot(mean, hist=False)
```
