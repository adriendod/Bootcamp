```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
%%time
df = pd.read_csv("./data/sales_predictions.csv")
```

```python
df.head(10)
```

```python
df.count()
```

```python
fig, ax =plt.subplots(figsize=(20,6))
sns.relplot(x="date", y="item_price", data=df.sample(50), ax = ax, kind="line")
```

```python
df_sample = df.sample(50, random_state = 42)
df_sample["date"] = pd.to_datetime(df_sample["date"])
```

```python
fig, ax =plt.subplots(figsize=(20,6))
sns.lineplot(x="date", y="item_price", data=df_sample, ax=ax)
```
