```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("./data/house_pricing.csv")
```

```python
df.info()
```

```python
fig, ax =plt.subplots(1,2, figsize=(20,6))
sns.relplot(x="LotArea", y="SalePrice", data = df,ax=ax[0])
sns.relplot(x="LotArea", y="SalePrice", data = df[(df["LotArea"] < 20000) & (df["SalePrice"] < 500000)],ax=ax[1])
plt.close()
plt.close()
```

```python
sns.relplot(x="LotArea", y="LotFrontage", data = df[(df["LotArea"] < 100000) & (df["LotFrontage"] < 200)])
```
